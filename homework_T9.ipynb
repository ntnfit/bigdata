```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
result = pd.read_excel('kq.xlsx')
result.info()
```

```

result.drop(index=[0],axis=1,inplace=True)
```

```
result
```

```
result['STT']=result['STT'].astype('int64')
```

```
scores=['Lý Thuyết','Win_Word','Excel','Powerpoint','Lý Thuyết TK','Thực Hành TK']
result=result[scores]
```

```

sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(
    result['Lý Thuyết TK'], norm_hist=False, kde=False, bins=20, hist_kws={"alpha": 1}
).set(xlabel='Lý Thuyết TK', ylabel='Count');
```

```

sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(
    result['Thực Hành TK'], norm_hist=False, kde=False, bins=20, hist_kws={"alpha": 1}
).set(xlabel='Thực Hành TK', ylabel='Count');
```
